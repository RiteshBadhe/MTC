2D Graph Plotting

Write a Python program to plot 2D graph of the function f(x) = e-x2 in [5,5] with green dashed points line with upwardpointing triangle

```python
from pylab import* 
import numpy as np 
x=np.linspace(-5,5,100) 
y=np.exp(-x**2)
plot(x,y,"-.^g",label="$y=e^{-x^2}$")
xlabel('x')
ylabel('y') 
title('y=e^{-x^2}') 
legend()
show()
```

3D Graph Plotting

```python
from mpl_toolkits import mplot3d
import numpy as np 
from pylab import* 
def f(x,y):
    return np.cos(x**2+y**2-0.5) 
    x=np.linspace(-1,1,1000)
    y=np.linspace(-1,1,1000)
    X,Y=np.meshgrid(x,y)
    Z=f(X,Y) 
ax=axes(projection='3d') 
ax.contour3D(X,Y,Z,50)
xlabel('x') 
ylabel('y') 
title('cos(x2 + y2 0.5)')
show()
```
